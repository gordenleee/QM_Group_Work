```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#read csv
data = pd.read_csv("./processed_data.csv")
```

```python
#Gcons2019

#specifies the columns in which the box plot is drawn
column_name = "Gcons2019"
column_data = data[column_name]

#outliers
flierprops = dict(marker='o', markerfacecolor='blue', markersize=1, markeredgecolor='none',linestyle='none')

#boxplot
plt.boxplot(column_data, vert=True, whis=3.1, flierprops=flierprops)

#set the title and axis labels
plt.title("Boxplot of {}".format(column_name))
plt.xlabel("{}".format(column_name))
plt.ylabel("Annual gas consumption/ kWh")

#display quartile
quartiles = [column_data.quantile(q) for q in [0.25, 0.5, 0.75]]
plt.text(1.16, quartiles[0], "Q1", va='center', ha='center')
plt.text(1.16, quartiles[1], "Q2", va='center', ha='center')
plt.text(1.16, quartiles[2], "Q3", va='center', ha='center')

#show plot
plt.show()
```

```python
#Gcons2005-2019

# select the column
columns_to_plot = ["Gcons2005", "Gcons2006", "Gcons2007", "Gcons2008", "Gcons2009", "Gcons2010", "Gcons2011", "Gcons2012", "Gcons2013", "Gcons2014", "Gcons2015", "Gcons2016", "Gcons2017", "Gcons2018", "Gcons2019"]

# create the DataFrame
data_to_plot = data[columns_to_plot]

# outliers
flierprops = dict(marker='o', markerfacecolor='blue', markersize=1, markeredgecolor='none', linestyle='none')

# boxplot
plt.boxplot(data_to_plot.values, labels=[col.replace("Gcons", "") for col in columns_to_plot], vert=True, whis=3, flierprops=flierprops)

# set the title and axis labels
plt.title("Boxplot of Gcons2005-2019")
plt.xlabel("Gcons2005-2019")
plt.ylabel("Annual gas consumption/ kWh")
plt.xticks(rotation=45)

# show plot
plt.show()
```

```python

# Set the figure size to accommodate three columns
plt.figure(figsize=(12, 8))

# Calculate and display statistics for each year
for i, year in enumerate(columns_to_plot):
    statistics_text = (
        f"{year}:\n"
        f"Mean: {np.mean(data[year]):.2f}\n"
        f"Median: {np.median(data[year]):.2f}\n"
        f"Mode: {int(np.argmax(np.bincount(data[year].astype(int))))}\n"
        f"Q1: {np.percentile(data[year], 25):.2f}\n"
        f"Q3: {np.percentile(data[year], 75):.2f}"
    )
    
    # Calculate the position for each year
    row = i % 3
    col = i // 3

    # Adjust the position parameters to reduce spacing
    plt.text( 0.18 * col, 0.8 - 0.18 * row, statistics_text, transform=plt.gca().transAxes, bbox=dict(facecolor='white', alpha=0.5), fontsize=8)

# Remove x and y axis labels
plt.xticks([])
plt.yticks([])

# Remove the outer box
plt.axis('off')

# Show the plot
plt.show()
```
