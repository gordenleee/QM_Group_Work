```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read csv
data = pd.read_csv("./processed_data.csv")
```

```python
#Gcons2019

# select the column
column_to_plot = "Gcons2019"
data_to_plot = data[column_to_plot]

# histograms
plt.hist(data_to_plot, bins=20, color='blue', edgecolor='black')  # 调整 bins 的数量以适应你的数据分布

# set the title and axis labels
plt.title(f"Histogram of {column_to_plot}")
plt.xlabel("Annual gas consumption/ kWh")
plt.ylabel("Frequency")

# show plot
plt.show()
```

```python
#Gcons2005-2019

# select the column
columns_to_plot = ["Gcons2005", "Gcons2006", "Gcons2007", "Gcons2008", "Gcons2009", "Gcons2010", "Gcons2011", "Gcons2012", "Gcons2013", "Gcons2014", "Gcons2015", "Gcons2016", "Gcons2017", "Gcons2018", "Gcons2019"]

# sreate plot
fig, axs = plt.subplots(len(columns_to_plot), 1, figsize=(10, 16), sharex=True, sharey=False)

# histograms
for i, column in enumerate(columns_to_plot):
    data_to_plot = data[column]
    bins = 20  
    
    axs[i].hist(data_to_plot, bins=bins, color=f'C{i}', edgecolor="black", alpha=0.5, label=column)
  
    # add legends for the small figures
    axs[i].legend()

# set the title and axis labels
plt.suptitle("Histograms of Gcons2005-2019", y=0.95)
plt.xlabel("Annual gas consumption/ kWh")
plt.ylabel("Frequency")

# adjust the layout
plt.tight_layout(rect=[0, 0, 0.9, 0.95])  # leave space for the years

# show plot
plt.show()
```
