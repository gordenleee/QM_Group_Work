```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read csv
data = pd.read_csv("./processed_data.csv")
```

```python
#PROP_TYPE Mid terrace
# select the column
columns_to_plot = ["Gcons2005", "Gcons2006", "Gcons2007", "Gcons2008", "Gcons2009", "Gcons2010", "Gcons2011",
                   "Gcons2012", "Gcons2013", "Gcons2014", "Gcons2015", "Gcons2016", "Gcons2017", "Gcons2018", "Gcons2019"]

# select data
selected_data = data[data["PROP_TYPE_Mid terrace"].isin([1])]

# sreate plot
fig, axs = plt.subplots(len(columns_to_plot), 1, figsize=(10, 16), sharex=True)

# histograms
for i, column in enumerate(columns_to_plot):
    data_to_plot = selected_data[column]
    bins = 20  
    
    axs[i].hist(data_to_plot, bins=bins, color=f'C{i}', edgecolor="black", alpha=0.5, label=column)
  
    # add legends for the small figures
    axs[i].legend()

# set the title and axis labels
plt.suptitle("Histograms of Gcons2005-2019 (PROP_TYPE=Mid terrace)", y=0.95)
plt.xlabel("Annual gas consumption (kWh)")
plt.ylabel("Frequency")

# adjust the layout
plt.tight_layout(rect=[0, 0, 0.9, 0.95])  # leave space for the years

# show plot
plt.show()
```
