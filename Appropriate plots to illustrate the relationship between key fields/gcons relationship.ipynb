```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read csv
data = pd.read_csv("./processed_data.csv")
```

```python
#PROP_TYPE

# select PROP_TYPE
prop_types = data['PROP_TYPE'].unique()

# sreate plot
fig, axes = plt.subplots(nrows=len(prop_types), ncols=1, figsize=(8, 8), sharex=True)

# histograms
for prop_type, ax in zip(prop_types, axes):
    subset = data[data['PROP_TYPE'] == prop_type]
    subset['Gcons2019'].plot(kind='hist', bins=20, color='steelblue', alpha=0.7, ax=ax, label=prop_type)

    ax.legend()  # add legends for the small figures

# set the title and axis labels
plt.suptitle("Relationship between PROP_TYPE and Gcons2019", y=1)
plt.xlabel("Annual gas consumption/ kWh")
plt.ylabel("Frequency")

# adjust the layout
plt.tight_layout()
plt.show()
```
